```python
class Money:
    def __init__(self, rubles: int, cents: int):
        self.__rubles = rubles
        self.__cents = cents

    # Decorator and getter for the amount of cents
    @property
    def cents(self):
        return self.__cents

    # Decorator and setter for the amount of cents
    @cents.setter
    def cents(self, cents):
        if cents >= 0:
            self.__cents = cents
        else:
            print('Да не оскудеет рука подающего, а не забирающего')


    # Decorator and getter for the amount of rubles
    @property
    def rubles(self):
        return self.__rubles

    # Decorator and setter for the amount of rubles
    @rubles.setter
    def rubles(self, rubles):
        if rubles >= 0:
            self.__rubles = rubles
        else:
            print('Да не оскудеет рука подающего, а не забирающего')


    def print_representation(self, counts=0):
        if self.__cents > 99:
            while self.__cents > 99:
                self.__cents -= 100
                counts += 1
            self.__rubles += counts
        print(f'There is {self.__rubles},{self.__cents} in my wallet')


    #Adition of money
    def addition(self, other_rubles=0, other_cents=0):
        rubles_add = int((((self.__rubles + other_rubles) * 100 + (self.__cents + other_cents))) // 100)
        cents_add = int((((self.__rubles + other_rubles) * 100 + (self.__cents + other_cents))) % 100)
        return print(f'There is {rubles_add},{cents_add} for each')


    # Division of money
    def fareshare(self, n_of_people):
        rubles_per_person = int(((self.__rubles * 100 + self.__cents) / n_of_people) // 100)
        cents_per_person = int(((self.__rubles * 100 + self.__cents) / n_of_people) % 100)
        return print(f'There is {rubles_per_person},{cents_per_person} for each')


    #Multiplication of money. Pay attention it could be multtplied and shared.
    def multy(self, coefficient, n_of_people=1):
        rubles_per_person = int(((self.__rubles * 100 + self.__cents)
                                 / n_of_people * coefficient) // 100)
        cents_per_person = int(((self.__rubles * 100 + self.__cents)
                                / n_of_people * coefficient) % 100)
        return print(f'There is {rubles_per_person},{cents_per_person} rn')


    # Comparison of wallets.
    def comparison(self, other_rubles: int, other_cents: int):
        rubles_comp = abs(int(((self.__rubles * 100 + self.__cents)
                               - (other_rubles * 100 + other_cents)) // 100))
        cents_comp = abs(int(((self.__rubles * 100 + self.__cents)
                              - (other_rubles * 100 + other_cents)) % 100))
        if (other_rubles * 100 + other_cents) > (self.__rubles * 100 + self.__cents):
            return print(f'Scrooge, you are richer than me by {rubles_comp},{cents_comp}')
        else:
            return print(f'I am richer than you by {rubles_comp},{cents_comp}')


wallet = Money(12, 550)
wallet.rubles = 100
wallet.cents = 50
wallet.fareshare(2)
wallet.multy(5.13)
wallet.comparison(99, 50)
wallet.addition(1, 2)
```
