```python
#Now we can estimate the dependence of survivals on Pclass, Sex, and Age 
df_interesting = df_exist[['Pclass', 'Sex', 'Age', 'Survived']]
df_interesting.head(10)
```

```python
metric1 = df_interesting[['Survived', 'Pclass']].groupby(['Pclass']).value_counts()
metric1
```

```python
metric2 = df_interesting[['Pclass', 'Sex', 'Survived']].groupby(['Sex', 'Pclass']).sum()
metric2
```

```python
#Sorting children out:
df_ch = df_interesting[df_interesting['Age'] <= 18] 
metric3 = df_ch[['Pclass', 'Survived']].groupby(['Pclass']).sum()
metric3
```

```python
#Find spouses, parent-child, siblings (SibSp - Number of Siblings/Spouses Aboard, Parch - Number of Parents/Children Aboard)  
```

```python
df_relationship = df_exist[['Name', 'Sex', 'Age', 'SibSp', 'Parch']]
names_extended = {0: 'Surname', 1: 'Name'}
df_addition = df_relationship['Name'].str.split(',',expand=True).rename(columns = names_extended).fillna('NaN')
df_relationship.insert(loc = 0 , column = 'Surname', value = df_addition[['Surname']])
df_relationship = df_relationship[['Surname', 'Sex', 'Age', 'SibSp', 'Parch']]
df_relationship
```

```python
#Defining spouses
df_spouses = df_relationship[df_relationship['SibSp'] == 1]
df_spouses =  df_spouses[df_spouses['Age'] >= 18]
metric5 = df_spouses[['Surname', 'Sex', 'Age']] .groupby(['Surname']).value_counts().to_frame().head(50)
metric5
#We need to sort this dataframe by the occurency of surnames: female+male
```

```python
#Defining parents-children
df_parents = df_relationship[df_relationship['Parch'] == 1]
df_parents.groupby(['Surname']).value_counts().to_frame().head(20)
```
