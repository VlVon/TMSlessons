```python
#We need to create the matrix:
import cv2
def blur(m_range, sigma):
    #Fill the expected matrix with zeros
    matrix = np.zeros((m_range, m_range))
    #Define the center coordinate
    c_coord = m_range // 2 
    #Fill each cell with the coordinate-dependent number
    for x in range(m_range):
        for y in range(m_range):
            matrix[x, y] = (1 / (2 * np.pi * sigma ** 2) * np.exp(((y - c_coord) ** 2 - (x - c_coord) ** 2)) / (2 * sigma ** 2))
    #The sum should be equal to 1 in order not to decrease the brigtness
    matrix /= np.sum(matrix)
    blurred_image = cv2.filter2D(img, -1, matrix)
    return matrix, plt.imshow(blurred_image)
print(blur(9, 1))
```
