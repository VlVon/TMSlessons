```python
import matplotlib.pyplot as plt
img = plt.imread("zhivotnoe-kot-bolshie-glaza.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
data = img.copy()
```

```python
data
```

```python
inverted_data = []
for chan_pack in data:
    inverted_chan_pack = []
    for channel in chan_pack:
        inverted_channel = [255 - r_g_b for r_g_b in channel]
        inverted_chan_pack.append(inverted_channel)
    inverted_data.append(inverted_chan_pack)
```

```python
plt.imshow(inverted_data)
```
